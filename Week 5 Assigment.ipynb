1.  Create a line plot of the provided dataset with years on the x-axis and population on the y-axis.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Year': [2000, 2005, 2010, 2015, 2020],
    'Population': [6.12, 6.45, 6.92, 7.35, 7.79]
}

df = pd.DataFrame(data)

plt.figure(figsize=(8,5))
plt.plot(df['Year'], df['Population'], marker='o', color='b', linestyle='-', linewidth=2 )
plt.title('Year vs Population Growth', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.xticks(df['Year'])         
plt.ylabel('Population (in billions)', fontsize=12)

plt.show()
```

6. Create a stacked bar chart of the provided dataset with the years on the x-axis and the total number of medals won by each country on the y-axis. The stacked bars should show the number of gold, silver, and bronze medals won.

10. Create a 3D scatter plot of the provided dataset with horsepower on the x-axis, weight on the y-axis, and miles per gallon on the z-axis.

```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D #found this on https://www.geeksforgeeks.org/3d-scatter-plotting-in-python-using-matplotlib/
df = pd.read_csv('C:\\Users\\Jeffm\\OneDrive\\DATA 747\\Week 5\\car_data.csv')
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['Horsepower'], df['Weight'], df['Mileage'], c='b', marker='o')
ax.set_xlabel('Horsepower')
ax.set_ylabel('Weight (lbs)')
ax.set_zlabel('Miles per Gallon (MPG)')
plt.title('3D Scatter Plot of Horsepower, Weight, and Mileage', fontsize=16)
plt.show()
```
