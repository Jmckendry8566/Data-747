### Dataset
The dataset to be used for this lab assignment is the Titanic dataset, which contains information about passengers on the Titanic, including their demographics, ticket information, and survival status. The dataset can be downloaded from Kaggle: https://www.kaggle.com/c/titanic/data

You can load the dataset using the following code:

Student Submission:

Calculate the summary statistics of the dataset and display them.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn import datasets as datasets
from ydata_profiling import ProfileReport

df = pd.DataFrame(titanic)
df.describe(include='all')
```

Visualize the distribution of the target variable using a bar chart.

```python
sns.countplot(x='survived', data=df)
plt.title('Distribution of Survival')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
```

Visualize the distribution of the target variable by class using a stacked bar chart

```python
survival_class = pd.crosstab(df['pclass'], df['survived'])
survival_class.plot(kind='bar', stacked=True)
plt.title('Survival by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.legend(title='Survived')
plt.show()
```

Plot a histogram to visualize the distribution of age

```python
sns.histplot(df['age'], bins=20, kde=True)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

Create a scatter plot to visualize the relationship between age and fare.

```python
plt.scatter(df['age'], df['fare'], alpha=0.5)
plt.title('Age vs Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

Use Pandas profiling to generate a detailed report of the dataset.

```python
profile = ProfileReport(df, title="Titanic Dataset Profiling Report", explorative = True)
profile.to_widgets()
```

Identify and visualize any missing values in the datase

```python
sns.heatmap(df.isnull(), cbar=False)
plt.title('Missing Values in the Dataset')
plt.show()
```

Identify and visualize any outliers in the dataset

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 5))
sns.boxplot(y=df['age'], ax=axs[0])
axs[0].set_title('Outliers in Age')
sns.boxplot(y=df['fare'], ax=axs[1])
axs[1].set_title('Outliers in Fare')
plt.show()
```
